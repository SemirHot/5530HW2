```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('/content/train.csv')
df.head()
```

```python
df = df.dropna(how='any')
df = df.reset_index(drop=True)
df.to_csv('cleanset.csv', index=False)
```

```python
df= pd.read_csv('/content/cleanset.csv')
df
```

```python
df['Mileage'] = df['Mileage'].str.replace(' kmpl', '', regex=False)
df['Mileage'] = df['Mileage'].str.replace(' km/kg', '', regex=False)
df['Engine'] = df['Engine'].str.replace(' CC', '', regex=False)
df['Power'] = df['Power'].str.replace(' bhp', '', regex=False)
df['New_Price'] = df['New_Price'].str.replace(' Lakh', '', regex=False)
df['New_Price'] = df['New_Price'].str.replace(' Cr', '', regex=False)
df.to_csv('cleanset.csv', index=False)
```

```python
df
```

```python
df_encoded = pd.get_dummies(df, columns=["Transmission", "Fuel_Type"])
df_encoded.to_csv('onehotencode.csv', index=False)
df_encoded
```

```python
df= pd.read_csv('/content/cleanset.csv')
df['New_Price'] = df['New_Price'].astype(float)
df['Price'] = df['Price'].astype(float)
df['Price_Chng'] = ((df['New_Price']) - (df['Price'])).round(2)
df['Current_Age'] = (2023) - (df['Year'])
df['Gas_Money'] = ((1.16) * ((df['Kilometers_Driven'] / df['Mileage'] ))).round(2)
df['CC_to_L'] = (df['Engine']) / (1000)
df.to_csv('manipulatedset.csv', index=False)
df
```
